```python
boolMap = {"Yes" : 1, "No" : 0}
df["International plan"] = df["International plan"].map(boolMap)
df["Voice mail plan"] = df["Voice mail plan"].map(lambda x : 1 if x == "Yes" else 0)
df.head()
```

```python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]
df.groupby(['Churn'])[columns_to_show].describe(percentiles=[])
```

```python
df.groupby(['Churn'])[columns_to_show].agg([np.mean, np.min, np.max])
```

```python
pd.crosstab(df['Churn'], df['International plan'])
```

```python
pd.crosstab(df['Churn'], df['International plan'], normalize=True, margins=True)
```
